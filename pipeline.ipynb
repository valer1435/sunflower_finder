```python
import numpy as np
import os
from matplotlib.image import imread
#Загрузка картинок в csv
plants = open("plants.csv", "w+")
sunflowers_dic = "sunflowerFieldsContours"
nosunflowers_dic = "nosunflowerFieldsContours"
max_ex = 20433
cnt = 0
# Загрузка подсолнухов
for image in os.listdir(sunflowers_dic):
    img =imread(sunflowers_dic+os.sep+image)
    img = np.reshape(img, (1, img.shape[0]*img.shape[1]*3))
    img = np.append(img, 1)
    row = str(img[0])
    for i in img[1::]:
        row+=" "+str(i)
    plants.write(row)
    plants.write("\n")
    cnt+=1
    if cnt >=20433:
        break
```

```python
# Загрузка Не подсолнухов
for image in os.listdir(nosunflowers_dic):
    img =imread(nosunflowers_dic+os.sep+image)
    img = np.reshape(img, (1, img.shape[0]*img.shape[1]*3))
    img = np.append(img, 0)
    row = str(img[0])
    for i in img[1::]:
        row+=" "+str(i)
    plants.write(row)
    plants.write("\n")
plants.close()
```

```python
import pandas as pd
data = pd.read_csv("plants.csv", sep=" ", header=None)
```

```python
data = data.sample(frac=1, replace=True, random_state=1)
data
```

```python
import torch


# determine the supported device

def get_device():
    if torch.cuda.is_available():
        device = torch.device('cuda:0')
    else:
        device = torch.device('cpu') # don't have GPU 
    return device

# convert a df to tensor to be used in pytorch
def df_to_tensor(df):
    device = get_device()
    return torch.from_numpy(df.values).float().to(device)
params = data.iloc[:, 0:-1].copy()

target = data.iloc[:, -1].copy()
```

```python
params = df_to_tensor(params)
target = df_to_tensor(target)
dataset_length = params.size()[0]
print((dataset_length* 0.60)//1)
train_x = params[: int(dataset_length* 0.60)]
valid_x = params[int(dataset_length* 0.60): int(dataset_length* 0.80)]
test_x = params[int(dataset_length* 0.80):]

train_y = target[: int(dataset_length* 0.60)]
valid_y = target[int(dataset_length* 0.60): int(dataset_length* 0.80)]
test_y = target[int(dataset_length* 0.80):]
```

```python
import torch.utils.data
batch_size = 64

train_tensor = torch.utils.data.TensorDataset(train_x, train_y) 
train_loader = torch.utils.data.DataLoader(dataset = train_tensor, batch_size = batch_size, shuffle = True)

INPUT_SIZE = train_x.size()[1]
OUTPUT_SIZE = 1
```

```python
import torch.nn as nn
import torch.nn.functional as F

class SunflowerFullConnectedClassificator(nn.Module):
    def __init__(self, input_size, output_size):
        super(SunflowerFullConnectedClassificator, self).__init__()
        self.fc1 = nn.Linear(input_size, 200)
        self.fc2 = nn.Linear(200, 200)
        self.fc3 = nn.Linear(200, output_size)

        
        
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        return self.fc3(x)
```

```python
import torch.optim as optim
learning_rate = 0.003
epochs = 10


net = SunflowerFullConnectedClassificator(INPUT_SIZE, OUTPUT_SIZE)
optimizer = optim.SGD(net.parameters(), lr=learning_rate, momentum=0.9, weight_decay=0.05)
criterion = nn.BCEWithLogitsLoss()
loss_arr = []
m = nn.Sigmoid()
for epoch in range(1, epochs+1):
    run_loss = 0
    cnt = 0
    for data in train_loader:
        X, y = data
        y = torch.reshape(y, (y.size()[0], 1))
        optimizer.zero_grad()
        outputs = net(X)
        loss = criterion(outputs, y)
        loss.backward()
        run_loss += loss
        #print(loss)
        optimizer.step()
        cnt+=1
    if epoch % 1 == 0:
        print(cnt)
        run_loss = run_loss / cnt
        loss_arr.append(run_loss)
        print(
            f"""-------------
            Epoch {epoch},
            epoch_loss = {run_loss}

        """)
        
        
print("Finished")
        
```

```python
loss_x = range(len(loss_arr))
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(loss_x, loss_arr)
ax.set_title("Train loss")
fig.show()
```

```python
m = nn.Sigmoid()
validation_outputs = net(valid_x)
valid_y = torch.reshape(valid_y, (valid_y.size()[0], 1))
valid_loss = criterion(validation_outputs, valid_y)
print("Valid loss:", valid_loss.item())
out = m(validation_outputs) >= 0.5
v = valid_y == out.float()
print("Num of true predicted:", torch.sum(v).item(), "from", valid_y.size()[0])


test_outputs = net(test_x)
test_y = torch.reshape(test_y, (test_y.size()[0], 1))
test_loss = criterion(test_outputs, test_y)
print("Test loss:", test_loss.item())
```

```python
import sklearn.metrics as metrics
test_outputs = m(net(test_x)) >= 0.5
print(metrics.accuracy_score(test_y, test_outputs))
```
